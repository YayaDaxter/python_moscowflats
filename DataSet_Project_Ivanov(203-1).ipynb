```python
df = pd.read_csv("flats_moscow.csv")
df.pop('code')
df.pop('Unnamed: 0')
print('-'*100)
```

```python
df
```

```python
df.info()
```

```python
df.describe(include='all')
```

Main description of dataset. There are no null values. 
***(2040 values in every column)***

```python
a = sns.relplot(x="totsp", y="price", hue='metrdist', data=df)
a.fig.set_figwidth(9)
a.fig.set_figheight(9)
```

The apartments cost depending on their total area square and metrdist 
**(The main distribution)**

```python
a = sns.relplot(x="totsp", y="price", kind="line", data=df)
a.fig.set_figwidth(9)
a.fig.set_figheight(9)
```

It is easy to observe the main tendency in cost depending on the change of total square.

```python
a = sns.relplot(x="totsp", y="kitsp", hue="price", kind="scatter", data=df)
a.fig.set_figwidth(9)
a.fig.set_figheight(9)
```

It is noticeable that the price is affected mostly by total square of a flat than kitchen square of the flat.

```python
a = sns.catplot(x="brick", y="price", kind="boxen",data=df)
a.fig.set_figwidth(9)
a.fig.set_figheight(9)
```

Prices for Aparments which were made of premium materials are much higher as the same from other materials. 

## Some statistics 

```python
df.pivot_table(['price'],['walk'], aggfunc='median')
```

Median flat price, depending on the possibility of walking without using any transport.

```python
print('Average distance from the apartment to the center:')
av_dist = df['dist'].median()
av_dist
```

```python
print('Comparison of average prices for apartments in different regions:')
print(round(df[df['dist'] < av_dist]['price'].mean(),2),';',round(df[df['dist'] > av_dist]['price'].mean(),2)) 
```

```python
print('Average apartment totsp:')
av_sp = df['totsp'].mean()
av_sp = round(av_sp)
av_sp
```

```python
print('Average apartment livesp and kitsp:')
l_sp = df['livesp'].mean()
l_sp = round(l_sp)
k_sp = df['kitsp'].mean()
k_sp = round(k_sp)
print(l_sp,';',k_sp)
```

```python
df['price'].std()
```

```python
print('Comparison of average prices for apartments on different floors:')
print(round(df[df['floor'] == 0]['price'].mean(),2),';',round(df[df['floor'] == 1]['price'].mean(),2)) 
```

```python
print('Comparison of average prices for apartments in different buildings:')
print(round(df[df['brick'] == 0]['price'].mean(),2),';',round(df[df['brick'] == 1]['price'].mean(),2)) 
```
